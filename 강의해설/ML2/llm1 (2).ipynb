```python
results = detector('/content/drive/MyDrive/data/도로.jpg')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
image_path = '/content/drive/MyDrive/data/도로.jpg'
image = plt.imread(image_path)
fig, ax = plt.subplots(1)
ax.imshow(image)

for result in results:
    xmin = int(result['box']['xmin'])
    ymin = int(result['box']['ymin'])
    xmax = int(result['box']['xmax'])
    ymax = int(result['box']['ymax'])
    width = xmax - xmin
    height = ymax - ymin
    score = result['score']
    label = result['label']

    rect = patches.Rectangle((xmin, ymin), width, height, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    ax.text(xmin, ymin, f'{label}: {score:.2f}', color='r', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

plt.axis('off')
plt.show()
```
